```python
# Dependencies and Setup
import pandas as pd
import json

# SQL Alchemy Dependencies
import sqlalchemy
from sqlalchemy import create_engine
```

```python
# Read movie_metadata.csv into movie_df
movie_df = pd.read_csv("Resources/movie_metadata.csv")
movie_df.rename(columns={'movie_title':'title'}, inplace=True)
```

```python
# Read all_script.csv into script_df
script_df = pd.read_csv("Resources/all_script.csv")
script_df
```

```python
movie_df.dtypes
```

```python
# Select
cleaned_movie_df = movie_df[["title","director_name","genres","duration","title_year","budget","content_rating","imdb_score","gross","actor_1_name","actor_2_name","facenumber_in_poster","country"]]
cleaned_movie_df
```

```python
cleaned_movie_df.genres = cleaned_movie_df.genres.str.split('|')
```

```python
cleaned_movie_df['title'].astype("string")
```

```python
script_df['title'].astype("string")
```

```python
script_df.dtypes
```

```python
movie_swear_df = pd.merge(cleaned_movie_df,script_df,how='inner',on='title')
#movie_df.join(script_df, on='title',how='outer')

movie_swear_df
```

```python
movie_swear_df.to_csv("Resources/movie_swear.csv")
```

```python
# Create database connection and load quake.df into database
rds_connection_string = "postgres:postgres@localhost:5432/movie_swear_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
cleaned_movie_df.to_sql(name='movie',con=engine, if_exists='replace',index=False)
```

```python
script_df.to_sql(name='script',con=engine, if_exists='replace',index=False)
```

```python
movie_swear_df.to_sql(name='movie_swear',con=engine, if_exists='replace',index=False)
```
