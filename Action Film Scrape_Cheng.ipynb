```python
movie_dict=[]

for i in range(len(test)): 
    movie_dict.append({'title':test[i],'Script':script_list[i], 'Swear_Count': [], 'Profanity':[], 'Fuck_Count':[],'Shit_Count':[],'Bitch_Count':[],"Genre":'Action'})
movie_dict[2]
```

```python
key_words=['fuck', 'shit','bitch','fucking','motherfucker','fucked','fucks','bitches','shitty']




for i in range(len(movie_dict)): 
    
    movie_key=movie_dict[i]
    movie_script=movie_key["Script"]
    fuck_counter=0
    shit_counter=0
    bitch_counter=0
    mother_fucking_counter=0
    for word in movie_script: 
            
        if word.lower() in key_words: 
            if (word.lower()=='fuck' or word.lower()=='fucking' or word.lower()=='fucks' or word.lower()=='fucked'): 
                fuck_counter+=1
                movie_key['Fuck_Count']=fuck_counter
                movie_key["Profanity"]=True
            elif (word.lower()=='shit' or word.lower()=='shitty'): 
                shit_counter+=1
                movie_key['Shit_Count']=shit_counter
                movie_key["Profanity"]=True
            elif (word.lower()=='bitch' or word.lower()=='bitches'):
                bitch_counter+=1
                movie_key['Bitch_Count']=bitch_counter
                movie_key["Profanity"]=True
            else:
                movie_key["Profanity"]=False
                

            
                
    print(movie_key['title'])
    print("------------------")
    print(f"The number of times fuck is said:{fuck_counter}")
    print(f"The number of times shit is said:{shit_counter}")
    print(f"The number of times bitch is said:{bitch_counter}")
    
    
    
    swear_counter=fuck_counter+shit_counter+bitch_counter
    movie_key["Swear_Count"]=swear_counter
    movie_key['Fuck_Count']=fuck_counter
    movie_key['Shit_Count']=shit_counter
    movie_key['Bitch_Count']=bitch_counter

    print(f'Profanity Total:{swear_counter}')
    print(f'Profanity: {movie_key["Profanity"]}')
    print("-----------------")

```
