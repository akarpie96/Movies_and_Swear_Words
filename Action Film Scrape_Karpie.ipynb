```python
from bs4 import BeautifulSoup
import requests
import pymongo
import pandas as pd
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import time 
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://imsdb.com/genre/War'
browser.visit(url)
html = browser.html
    # Parse HTML with Beautiful Soup
soup = BeautifulSoup(html, 'html.parser')
test=[]
titles=soup.find_all('p')
titles
# titles[0].find('a').text
for title in titles: 
    names=title.find('a').text
    test.append(names)
    


test
```

```python
url = 'https://imsdb.com/genre/War'
browser.visit(url)
script_list=[]
for name in test: 
    try: 
        browser.links.find_by_partial_text(name).click()
        browser.links.find_by_partial_text("Script").click()
        html = browser.html
            # Parse HTML with Beautiful Soup|
        soup = BeautifulSoup(html, 'html.parser')
        script=soup.find("td", class_="scrtext").find('pre').text.split()
        script_list.append(script)

        browser.links.find_by_partial_text("War").click()
    except: 
        print(f"No Script for:{name}")
        script_list.append("null")
        browser.visit(url)
```

```python
movie_dict=[]

for i in range(len(test)): 
    movie_dict.append({'title':test[i],'Script':script_list[i], 'Swear_Count': [], 'Profanity':[]})
movie_dict[31]
```

```python
key_words=['fuck', 'shit','bitch']


fuck_counter=0
shit_counter=0
bitch_counter=0
mother_fucking_counter=0

for i in range(len(movie_dict)): 
    movie_key=movie_dict[i]
    movie_script=movie_key["Script"]
    for word in movie_script: 
        if word in key_words: 
            if word=='fuck': 
                fuck_counter+=1
            elif word=='shit': 
                shit_counter+=1
            elif word=='bitch':
                bitch_counter+=1

    print(movie_key['title'])

    print(f"The number of times fuck is said:{fuck_counter}")
    print(f"The number of times shit is said:{shit_counter}")
    print(f"The number of times bitch is said:{bitch_counter}")

    swear_counter=fuck_counter+shit_counter+bitch_counter
    print(f'Profanity Total:{swear_counter}')

print(swear_counter)
```
