```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
url = 'https://imsdb.com/genre/War'
# browser.visit(url)
script_list=[]
for name in titles: 
    try: 
        url =f'https://imsdb.com/scripts/{name}.html'
        browser.visit(url)
#         browser.links.find_by_partial_text(name).click()
#         browser.links.find_by_partial_text("Script").click()
        html = browser.html
            # Parse HTML with Beautiful Soup|
        soup = BeautifulSoup(html, 'html.parser')
        script=soup.find("td", class_="scrtext").find('pre').text.split()
        script_list.append(script)

        browser.links.find_by_partial_text("War").click()
    except: 
        print(f"No Script for:{name}")
        script_list.append("null")
        browser.visit(url)
```

```python
movie_dict=[]

for i in range(len(test)): 
    movie_dict.append({'title':test[i],'Script':script_list[i], 'Swear_Count': [], 'Profanity':[], 'Fuck_Count':[],'Shit_Count':[],'Bitch_Count':[]})
movie_dict
```

```python
key_words=['fuck', 'shit','bitch']




for i in range(len(movie_dict)): 
    
    movie_key=movie_dict[i]
    movie_script=movie_key["Script"]
    fuck_counter=0
    shit_counter=0
    bitch_counter=0
    mother_fucking_counter=0
    for word in movie_script: 
            
        if word in key_words: 
            if word.lower()=='fuck': 
                fuck_counter+=1
                movie_key['Fuck_Count']=fuck_counter
                movie_key["Profanity"]=True
            elif word.lower()=='shit': 
                shit_counter+=1
                movie_key['Shit_Count']=shit_counter
                movie_key["Profanity"]=True
            elif word.lower()=='bitch':
                bitch_counter+=1
                movie_key['Bitch_Count']=bitch_counter
                movie_key["Profanity"]=True
            elif word not in key_words:
                movie_key["Profanity"]=False

            
                
    print(movie_key['title'])
    print("------------------")
    print(f"The number of times fuck is said:{fuck_counter}")
    print(f"The number of times shit is said:{shit_counter}")
    print(f"The number of times bitch is said:{bitch_counter}")

    swear_counter=fuck_counter+shit_counter+bitch_counter
    movie_key["Swear_Count"]=swear_counter
    
    print(f'Profanity Total:{swear_counter}')
    print("-----------------")

print(swear_counter)
```

```python
war_df=pd.DataFrame.from_dict(movie_dict)
war_df=war_df[["title",'Swear_Count',"Fuck_Count","Shit_Count","Bitch_Count","Profanity"]]
war_df.replace("", 0)
war_df["Profanity"].fillna(False)
war_df
```
